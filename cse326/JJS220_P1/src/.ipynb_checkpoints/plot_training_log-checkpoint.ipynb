```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pickle

with open('../data/batch_outcome.pkl', 'rb') as f:
    (w, b, log) = pickle.load(f)
```

```python
x = np.array(range(1,len(log)+1))
training_loss = [e[0] for e in log]
validation_loss = [e[1] for e in log]
test_loss = [e[2] for e in log]
# print (x)
# print (training_loss)
plt.figure()
plt.plot(x, training_loss, 'y-', label='training_loss')
plt.plot(x, validation_loss, 'b-', label='validation_loss')
plt.plot(x, test_loss, 'r-', label='test_loss')
plt.legend()

param_norm = [e[3] for e in log]
plt.figure()
plt.plot(x, param_norm)
```
